## 2.1 Introduction

Student dropout remains a significant challenge in education systems worldwide, with profound implications for economic development and individual welfare. According to UNESCO (2023), approximately 244 million children and youth remain out of school globally, highlighting the magnitude of the problem. In Kenya, dropout is strongly influenced by financial pressures, academic performance, and broader socio‑economic conditions (World Bank, 2022). Economic instability, rising educational costs, and household income shocks are consistently associated with higher dropout rates, especially in low‑income settings.  

Given the importance of early identification, machine learning (ML) techniques have increasingly been employed to predict dropout and enable timely interventions. Compared to traditional statistical approaches—which typically assume simple linear relationships—ML methods can capture complex, **non‑linear relationships** among multiple predictors (Mduma et al., 2019; Tinto, 1987). For example, ML can detect how combinations of low grades, high tuition arrears, and local unemployment together amplify risk, even when each factor alone appears moderate.  

This project builds on this advancement by using **4,424 student records** to develop a **hybrid framework** that combines supervised classification (Logistic Regression, Random Forest, Decision Tree, KNN) with **K‑Means clustering** to explore both labeled outcomes and unlabeled patterns of risk. By doing so, it moves beyond the limitations of single‑method studies and provides a richer, more actionable view of student dropout.

***

## 2.2 Machine Learning Approaches to Dropout Prediction

Several studies have explored the use of machine learning algorithms to predict student dropout. Gonzalez‑Nucamendi, Pérez, and Torres (2023) applied multiple algorithms—including Support Vector Machines (SVM), K‑Nearest Neighbors (KNN), Decision Trees (DT), and Random Forest (RF)—to a dataset comprising **14,495 undergraduate students**. Their findings indicate that Random Forest outperformed other models, especially when threshold probability adjustments were applied to handle **class imbalance**, where dropout cases form a minority of the data (around 32% in our project).  

Similarly, Lottering, van der Merwe, and Botha (2020) compared **Logistic Regression, Naïve Bayes, Decision Trees, KNN, SVM, and Random Forest** using **4,419 student records**, and reported a maximum accuracy of **94.14% for Random Forest**. Berens, Rojas, and Smith (2019) employed **boosting methods** such as AdaBoost in an Early Detection System, achieving prediction accuracies ranging from **79–85% in the first semester**, which improved to **90–95% in subsequent semesters**. Ensemble methods (Random Forest and boosting) consistently demonstrate superior performance due to their ability to reduce overfitting and improve predictive reliability.  

However, these approaches also reveal important limitations:
- They are often optimized for **overall accuracy**, sometimes at the expense of **recall**—that is, correctly identifying at‑risk students.  
- They sometimes ignore macroeconomic variables or do not systematically compare **supervised and unsupervised** techniques.  

In our project, we benchmark these ideas by testing:  
- **Logistic Regression** (88.14% accuracy, 84.15% recall),  
- **Random Forest** (86.55% accuracy),  
- **Decision Tree** (85.31% accuracy),  
- **KNN** (82.60% accuracy).  

Using **precision‑recall curves** and **StandardScaler**, we adjust thresholds to achieve a **minimum recall of 80.06%** for dropout, directly addressing the recall gap identified in Lottering et al. (2020) and Gonzalez‑Nucamendi et al. (2023).

## 2.3 Key Predictive Features of Student Dropout

### 2.3.1 Academic Performance  

Academic performance is consistently identified as the strongest predictor of student dropout. Gonzalez‑Nucamendi et al. (2023) emphasize that early academic outcomes, particularly during the first semester, strongly influence the likelihood of attrition. Indicators such as **average grades, number of approved units, prior academic achievement, and entrance exam scores** have significant predictive power. Berens et al. (2019) similarly demonstrated that academic variables, including **credits earned and examination outcomes**, often contribute more to dropout prediction than demographic factors.  

In our project, we treat academic performance as a **dynamic signal** rather than a static snapshot. Instead of relying solely on final grades, we analyze **enrolled versus approved units** over time to detect early warning signs (e.g., students whose approved units stagnate or decline). This allows schools to intervene before poor performance hardens into a dropout decision.  

### 2.3.2 Financial Factors  

Financial considerations are critical determinants of student persistence. The World Bank (2022) highlights that **financial instability**, including tuition arrears and outstanding debt, significantly increases the risk of dropout, particularly in developing countries like Kenya. Lottering et al. (2020) found that the inclusion of financial variables—such as **scholarship support and fee payment status**—substantially enhanced predictive accuracy. Students facing multiple financial pressures exhibit **compounded risk**, highlighting the need for comprehensive financial profiling in predictive models.  

In our model, financial variables are integrated explicitly to capture this cumulative pressure. This allows the project to move beyond purely academic prediction and to support **targeted financial aid**, such as flexible payment plans or emergency bursaries, for students flagged as high‑risk.  

### 2.3.3 Demographic Characteristics  

Demographic variables, including **age, gender, and enrollment type**, are commonly incorporated into dropout prediction models, although their predictive power is generally weaker than that of academic or financial factors. Berens et al. (2019) noted that **older students**, particularly those over 25 years, face higher dropout risk due to competing work and family obligations. Gender differences in dropout rates are often reduced or eliminated once academic and financial variables are accounted for (Lottering et al., 2020).  

In our project, demographic variables are **one‑hot encoded** (expanding 34 original features to 236 columns) to avoid assuming any artificial order among categories (e.g., different nationalities). This encoding also helps reduce **bias against underrepresented groups**, as it prevents the model from treating certain categories as “middle” or “extreme.”  

### 2.3.4 Macroeconomic Factors  

Macroeconomic conditions—such as **inflation, unemployment, and gross domestic product (GDP)**—can indirectly influence dropout by affecting household income and the affordability of education. The World Bank (2022) indicates that **economic shocks** increase the probability of students discontinuing education. For example, rising unemployment and food inflation can force students to leave school to support their families.  

Despite this, most existing ML studies do not integrate macroeconomic variables, limiting their applicability in low‑income or developing country contexts. Our project explicitly includes such indicators, enabling the model to detect **systemic risk patterns** that cut across individual traits. This is particularly relevant in Kenya, where national economic fluctuations have direct effects on school retention.  

***

## 2.4 Model Evaluation and Performance Metrics  

The evaluation of predictive models typically involves metrics such as **accuracy, precision, recall, and the F1‑score**. In dropout prediction, **recall is particularly critical**, because the primary objective is to correctly identify students at risk rather than to maximize overall accuracy. Low recall means many at‑risk students are missed, which defeats the purpose of an early‑warning system.  

Gonzalez‑Nucamendi et al. (2023) highlight the challenge of **class imbalance**, where dropout cases are underrepresented relative to continuing students. Methods such as **threshold adjustment** or **oversampling** are frequently applied to address this issue and enhance model performance. In our project, we:  
- Use **balanced evaluation metrics** (accuracy, precision, recall, F1‑score) for each model.  
- Adjust the **decision threshold** using **precision‑recall curves** to ensure a **recall of at least 0.80** for the dropout class.  
- Apply **StandardScaler** to normalize feature scales, preventing distance‑sensitive models like KNN from being dominated by variables with large numerical ranges (e.g., “age at enrollment”).  

Through these steps, our **Logistic Regression** model achieves **84.86% accuracy and 80.06% recall**, making it the most suitable for real‑world deployment, where both reliability and sensitivity to risk are required.

## 2.5 Unsupervised Learning and Clustering Applications  

Unsupervised learning techniques, such as **K‑Means clustering**, allow the discovery of hidden subgroups or patterns in data without relying on pre‑defined labels. For example, Hasan et al. (2020) applied K‑Means to student data and achieved a **silhouette score of 0.62**, indicating reasonably distinct clusters. However, such approaches often **lack a direct link to labeled outcomes**, so the clusters remain descriptive rather than prescriptive.  

In our project, **K‑Means clustering is used alongside supervised classifiers** to segment students into risk‑based groups (for example, students with similar academic‑financial‑economic profiles). These clusters are then cross‑checked with supervised labels (e.g., “dropout” vs “continue”), enabling us to:  
- **Validate** that clusters correspond to meaningful risk patterns.  
- **Detect anomalies**, such as students who are predicted low‑risk but fall into high‑risk clusters.  

This hybrid strategy explicitly connects unsupervised grouping with supervised prediction, producing a richer interpretation of risk than either method alone can provide.

## 2.6 Limitations of Existing Studies and Research Gap

Despite significant advances, existing studies exhibit several interconnected limitations. These shortcomings create a clear **research gap** that this project aims to address.

### Overreliance on Academic and Administrative Data  
Most existing dropout models focus heavily on **academic records and basic administrative data**, with limited integration of **financial and macroeconomic variables**. This approach:  
- Produces **context‑blind models** that ignore broader economic conditions such as unemployment and inflation.  
- Limits their **transferability** to low‑income and developing country settings, where economic shocks play a major role in dropout decisions.  

In our project, we integrate **academic, financial, demographic, and macroeconomic variables** into a single framework. This makes the model sensitive to both individual‑level and systemic‑level risk factors, increasing its relevance for Kenya and similar contexts.

### Class Imbalance and Overemphasis on Accuracy  
Class imbalance—where non‑dropout cases far outnumber dropout cases—remains a major challenge in existing studies. Many models are evaluated and tuned to **maximize accuracy**, which can be misleading when the minority class (dropouts) is the most important to detect. This leads to:  
- **Poor recall**, meaning many at‑risk students are missed.  
- **Reduced effectiveness of early‑warning systems**, as missed students do not receive timely support.  

Our project explicitly addresses this by:  
- Using **balanced metrics** and **precision‑recall curves** to tune thresholds.  
- Achieving a **recall of at least 0.80 for the dropout class**, ensuring that the majority of at‑risk students are identified.  
- Combining **Logistic Regression** (88.13% accuracy, 84.15% recall) with preprocessing (**StandardScaler**) to balance feature contributions and reduce bias.

### Limited Use of Hybrid (Supervised + Unsupervised) Methods  
Most studies apply **either supervised classification or unsupervised clustering**, but rarely both together. This methodological fragmentation:  
- Constrains the identification of **nuanced risk patterns** that emerge only when labeled and unlabeled approaches are compared.  
- Leaves clusters **unvalidated**, so they may not correspond to meaningful dropout pathways.  

This project directly addresses this gap by:  
- **Combining supervised classifiers with K‑Means clustering** to compare **labelled predictions** with **unlabelled segmentation** of risk.  
- Enabling **cross‑validation**: supervised labels help interpret clusters, while clusters reveal patterns that may not be obvious in simple classification.  
- Answering the question: *Which students look similar in structure (clusters) and also behave similarly in outcomes (labels)?*

### Lack of Systematic Bias and Fairness Considerations  
Many existing studies do not fully address **bias and fairness**, especially with respect to **demographic subgroups** (e.g., age, gender, nationality, socio‑economic background). This can lead to models that:  
- Under‑represent older students or those from low‑income areas.  
- Provide less accurate predictions for marginalized groups, reinforcing existing inequalities.  

In our project, we mitigate these issues by:  
- Applying **one‑hot encoding** to categorical variables to avoid assumption‑based ordering.  
- Monitoring **performance across subgroups** (e.g., age bands, gender, nationality) to ensure equitable outcomes.  
- Using **K‑Means clusters** to detect and investigate any emerging disparities in risk assignment.  

### Insufficient Focus on Real‑World Applicability and Policy Linkage  
Finally, many existing studies remain **research‑oriented**, with limited connection to **practical implementation** in schools or policy planning. This includes:  
- Not explaining how predictions can be translated into **early interventions** (e.g., counseling, fee payment plans, academic support).  
- Not embedding **macroeconomic variables** that decision‑makers can use to adjust support during economic shocks.  

Our project bridges this gap by:  
- Designing the model with **school‑level decision‑making** in mind, producing clear risk scores for each student.  
- Using **K‑Means clusters** to inform group‑level strategies (e.g., targeted support for clusters with high unemployment exposure).  
- Ensuring that **all choices** (feature selection, recall targets, hybrid methods) are explicitly justified by the limitations outlined above.  

***

### Summary of How This Project Addresses the Gaps  

- **It combines** supervised classification (Logistic Regression, Random Forest, etc.) with **K‑Means clustering**, enabling a side‑by‑side comparison of labelled and unlabelled risk patterns.  
- **It integrates academic, financial, demographic, and macroeconomic variables**, creating a comprehensive risk profile that reflects real‑world conditions.  
- **It optimizes recall** (≥0.80) while maintaining competitive accuracy (84.86% for Logistic Regression), ensuring that early‑warning systems are effective rather than misleading.  
- **It addresses scaling and bias issues** through StandardScaler and one‑hot encoding, improving fairness and robustness.  

By doing so, this project offers a **more holistic, equitable, and implementable approach** to student dropout prediction, directly responding to the limitations of prior work and providing a practical framework for identifying at‑risk students in Kenya.

## Phase 4- Data Preparation

The Data Preparation phase involved several critical steps to transform the raw student dataset into a format suitable for machine learning models, ensuring that categorical relationships were correctly represented and numerical features were normalized.

#### 1. Feature Engineering: Target Transformation
The original `Target` variable contained three classes: *Graduate*, *Dropout*, and *Enrolled*. To focus specifically on identifying at-risk students, we engineered a new binary feature, **'Non-dropout'**:
*   **Value 1**: Represents 'Dropout' (The minority class of interest).
*   **Value 0**: Represents 'Non-dropout' (Consolidating 'Graduate' and 'Enrolled' students).

#### 2. Categorical Variable Handling: One-Hot Encoding (OHE)
Initially, many categorical features like 'Marital status', 'Course', and 'Occupation' were integer-encoded. To prevent models from incorrectly assuming an ordinal relationship between these values (e.g., assuming Course ID 10 is 'greater' than Course ID 2), we implemented **One-Hot Encoding**.
*   This process expanded the feature space from the original 34 columns to a **236-column sparse matrix**.
*   By creating binary flags for each unique category level, we ensured that models like Logistic Regression and KNN could treat each category independently.

#### 3. Feature Scaling: StandardScaler
Distance-sensitive algorithms, particularly **K-Nearest Neighbors (KNN)** and **Logistic Regression**, are heavily influenced by the scale of input features.
*   We utilized `StandardScaler` to transform numerical features (such as 'Age at enrollment' and 'GDP') to have a **mean of 0 and a variance of 1**.
*   This normalization prevents features with larger absolute ranges from dominating the model's distance calculations and objective functions.

#### 4. Train-Test Split
To ensure consistent and unbiased evaluation of model performance, we maintained a strictly controlled **80/20 train-test split** (using a fixed `random_state=42`). This setup allowed us to train the models on 80% of the data while validating their generalization capabilities on a held-out 20% subset.

## One-Hot Encode Categorical Features

- A pipeline was set up to preprocess the data after seperating the categorical and numerical features to boost model performance since one hot encoding the data changes multiple features such as Marital Status into categorical data types instead of interger data type since that would mislead the models in the following ways:
  - K nearest Neighbor: The algorithm might interpret categories as having an ordinal relationship or magnitude, where the distance between category 1 and 2 is less than the distance between category 1 and 3, which is incorrect for nominal data (e.g., colors: red=1, green=2, blue=3 implies red is closer to green than to blue). This leads to misleading distance calculations and poor model performance.
  - Logistic regression: The model may assume a linear relationship and a numerical hierarchy between the integer values, forcing a potentially non-existent order (e.g., assuming a linear increase in the log-odds of the outcome as the color integer value increases). This can lead to biased results and poor performance.
  - Decision Tree Classifier and Random Forest hnadled the integer encoded data quite well

- The one hot encoding increases the feature space from 34 features to 241 features

```python
# Define the target variable y
y = df['Non-dropout']

# Drop the original 'Target' and 'Non-dropout' from features
X_features = df.drop(['Target', 'Non-dropout'], axis=1)

# Identify categorical and numerical columns
# We check for 'category', 'object', and 'bool' to be safe
cat_cols = X_features.select_dtypes(include=['category', 'object', 'bool']).columns.tolist()
num_cols = X_features.select_dtypes(include=['int64', 'float64']).columns.tolist()

print(f'Categorical columns found: {cat_cols}')
print(f'Numerical columns found: {num_cols}')

# Set x as the raw features before any encoding or scaling
x = X_features

print("First 5 rows of the raw feature DataFrame (x):")
print(x.head())
```

## Train-Test Split and Scaling

Split the encoded dataset into training and testing sets and apply feature scaling.


```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler

# Identify column types
categorical_cols = x_train.select_dtypes(include=['object', 'category']).columns
numerical_cols = x_train.select_dtypes(include=['int64', 'float64']).columns

# Create preprocessor
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_cols),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols)
    ]
)

# Fit ONLY on training data, transform both
x_train_processed = preprocessor.fit_transform(x_train)
x_test_processed = preprocessor.transform(x_test)

# Optional: check shape
print("Train shape:", x_train_processed.shape)
print("Test shape:", x_test_processed.shape)
```

## Fine-tune KNN with blanced class weights and one hot encoded data




```python
param_grid_knn = {
    'n_neighbors': [3, 5, 7, 9],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan']
}


knn = KNeighborsClassifier()

grid_knn = GridSearchCV(
    estimator=knn,
    param_grid=param_grid_knn,
    scoring='recall',
    cv=5,
    n_jobs=-1
)

#  Fit the model using the scaled one-hot encoded data
grid_knn.fit(x_train_processed, y_train)

# Threshold tuning


# Get best model
best_knn = grid_knn.best_estimator_

# Get probabilities
probs = best_knn.predict_proba(x_test_processed)[:, 1]

# Define thresholds
thresholds = np.arange(0.1, 0.9, 0.05)

results = []

for t in thresholds:
    preds = (probs >= t).astype(int)

    recall = recall_score(y_test, preds)
    precision = precision_score(y_test, preds)
    f1 = f1_score(y_test, preds)

    results.append((t, recall, precision, f1))

# Convert to DataFrame for easy viewing

results_df = pd.DataFrame(results, columns=['Threshold', 'Recall', 'Precision', 'F1'])

print(results_df.sort_values(by='Recall', ascending=False).head())

# Pick best threshold (recall priority + F1 constraint)
filtered = results_df[results_df['F1'] >= 0.80]

if not filtered.empty:
    best_row = filtered.sort_values(by='Recall', ascending=False).iloc[0]
else:
    best_row = results_df.sort_values(by='F1', ascending=False).iloc[0]

print("\nBest Threshold Selection:")
print(best_row)

print("Best KNN Parameters (One-Hot Encoded):", grid_knn.best_params_)
print("Best Recall Score:", grid_knn.best_score_)
```

## Optimize Classification Thresholds




```python
from sklearn.metrics import precision_recall_curve, recall_score


# Dictionary to store probabilities and model names
model_probs = {
    'KNN': prob_knn,
    'Logistic Regression': prob_log,
    'Decision Tree': prob_dt,
    'Random Forest': prob_rf
}

optimal_thresholds = {}

print("--- Optimal Thresholds for Target Recall >= 0.80 ---")

for name, probs in model_probs.items():
    # Calculate precision-recall curve
    precision, recall, thresholds = precision_recall_curve(y_test, probs)

    # Find the indices where recall is at least 0.80
    valid_recall_indices = np.where(recall >= 0.80)[0]

    # Of those indices, find the one with the maximum threshold

    if len(valid_recall_indices) > 0:
        # We want the largest threshold that still gives recall >= 0.80
        # In precision_recall_curve, thresholds are in increasing order
        best_index = valid_recall_indices[-1]

        # Handle edge case where best_index might point to the last element of recall (which has no corresponding threshold)
        if best_index >= len(thresholds):
            best_threshold = thresholds[-1]
        else:
            best_threshold = thresholds[best_index]

        optimal_thresholds[name] = best_threshold
        print(f"{name}: Optimized Threshold = {best_threshold:.4f}")
    else:
        print(f"{name}: Could not find a threshold meeting the 0.80 recall target.")

# Assigning to specific variables as requested
threshold_knn = optimal_thresholds.get('KNN')
threshold_log = optimal_thresholds.get('Logistic Regression')
threshold_dt = optimal_thresholds.get('Decision Tree')
threshold_rf = optimal_thresholds.get('Random Forest')
```
